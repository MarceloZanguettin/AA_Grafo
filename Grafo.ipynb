```python
class No:
    info=0
    proximo=None
    def __init__(self, info,proximo=None):
        self.info=info
        self.proximo=proximo
```

```python
class Grafo:
    vertices=None
    
    def __init__(self):
        self.vertices=[]
    
    def insereVertice(self, v):
        self.vertices.append(No(v))
        
    def insereAresta(self, v1, v2):
        no1=No(v1)
        no2=No(v2)
        for v in self.vertices:
            if v.info==v1:
                while(v.proximo!=None):
                    v=v.proximo
                v.proximo=no2
                break
        for v in self.vertices:
            if v.info==v2:
                while(v.proximo!=None):
                    v=v.proximo
                v.proximo=no1
                break
            
    def listaTodasArestas(self):
        for v in self.vertices:
            vertice=v.info
            while(v.proximo!=None):
                v=v.proximo
                print('Aresta:',vertice, ',',v.info)
    
    def verticesAdjacentes(self, vtc):
        for v in self.vertices:
            if v.info==vtc:
                print('Vertice(s) adjacente(s) a :', vtc)
                v=v.proximo
                while (v!=None):
                    print(str(v.info))
                    v=v.proximo
                break

    def grauVertice(self, vtc):
        quantidade=0;
        for v in self.vertices:
            if v.info==vtc:
                v=v.proximo
                while (v!=None):
                    v=v.proximo
                    quantidade+=1;
                print('A quantidade de arestas do vertice', vtc +' é igual a',quantidade)
                break

    def grauGrafo(self):
        grauGrafo=0;
        for v in self.vertices:
            grauVertice=0;
            v=v.proximo
            while (v!=None):
                v=v.proximo
                grauVertice+=1;
            if grauVertice>grauGrafo:
                grauGrafo=grauVertice;
        print('O grau do Grafo é:',grauGrafo)

    def verificaRegular(self):
        grauGrafo=0
        cont=0
        for v in self.vertices:
            grauVertices=0
            v=v.proximo
            while v!=None:
                grauVertices+=1
                v=v.proximo
            if grauGrafo==0:
                grauGrafo=grauVertices
            elif grauGrafo!=grauVertices:
                return False
                break
        return True

    def verificaAresta(self, v1, v2):
        for v in self.vertices:
            if v.info==v1:
                v=v.proximo
                while v!=None:
                    if v.info==v2:
                        return True
                    v=v.proximo
                break
        return False
        
    #Exercico 1
    def ordemGrafo(self):
        return len(self.vertices)

    #Exercicio 2
    def arestasGrafo(self):
        quantidade=0;
        for v in self.vertices:
            while(v.proximo!=None):
                v=v.proximo
                quantidade+=1
        return quantidade/2
    
    #Exercicio 3
    def verificaCompleto(self):
        t=len(self.vertices)
        for v in self.vertices:
            grauVertice=0;
            v=v.proximo
            while (v!=None):
                v=v.proximo
                grauVertice+=1;
            if grauVertice!=t:
                return False;
        return True

    #Exercicio 4
    def verificaDesconexo(self):
        for v in self.vertices:
            grauVertice=0;
            v=v.proximo
            while (v!=None):
                v=v.proximo
                grauVertice+=1
            if grauVertice!=0:
                return False
            else:
                return True
```

```python
g=Grafo()
g.insereVertice('a')
g.insereVertice('b')
g.insereVertice('c')
g.insereVertice('d')
g.insereVertice('e')
g.insereAresta('a','b')
g.insereAresta('a','e')
g.insereAresta('b','c')
g.insereAresta('b','e')
g.insereAresta('b','d')
g.insereAresta('c','d')
g.insereAresta('d','e')
g.listaTodasArestas()
g.verticesAdjacentes('a')
g.verticesAdjacentes('d')
g.grauVertice('a')
g.grauGrafo()
if g.verificaRegular():
    print('O grafo é regular!')
else:
    print('O grafo não é regular')
if g.verificaAresta('a','b'):
    print('Existe aresta entre os vertices')
else:
    print('Não existe aresta entre os vertices')
if g.verificaAresta('a','c'):
    print('Existe aresta entre os vertices')
else:
    print('Não existe aresta entre os vertices')
print('A ordem do grafo é:',g.ordemGrafo())
print('A quantidade de arestas no grafo é de:',g.arestasGrafo())
if g.verificaCompleto():
    print('O grafo é completo')
else:
    print('O grafo não é completo')
if g.verificaDesconexo():
    print('O grafo é Desconexo')
else:
    print('O grafo não é Desconexo')
```
